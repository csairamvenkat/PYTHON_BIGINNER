# Class Data

```python
RandomAddresses=pd.read_excel(r'C:\Users\sai.venkat\OneDrive\DataScience\Python\PYTHON_BIGINNER\Libraries\Pandas\Bangalore_Addresses.xlsx')
RandomAddresses
```

```python
Banglore_Address_Split=RandomAddresses['Address'].str.split(r'[,\s-]+',expand=True)
Banglore_Address_Split



```

```python
Banglore_Date_Split=RandomAddresses['Date'].str.split(r'[,\s-]+',expand=True)
Banglore_Date_Split
```

```python
Banglore_Time_Split=RandomAddresses['Time'].str.split(r'[,\s-]+',expand=True)
Banglore_Time_Split=Banglore_Time_Split.replace({'hr': ' Hours', 'min': ' Minutes', 'sec': ' Seconds'}, regex=True)
Banglore_Time_Split
```

# To CSV

```python
import pandas as pd


# Define the header (column names)
header = ['Building No.', 'Block Number ', 'Area', 'City','PinCode','Date','Month','Year','Hour','Minute','Sec']
UpdatedDf=pd.concat([Banglore_Address_Split,Banglore_Date_Split,Banglore_Time_Split],axis=1)
# Sample data (CleanedCOntactData)
UpdatedDfList = UpdatedDf.values.tolist()
# Create DataFrame with specified column names
df = pd.DataFrame(UpdatedDfList, columns=header)

# Write To Csv
df.to_csv('Delimitted Addresses.csv', index=False)

```
