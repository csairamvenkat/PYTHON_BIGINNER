```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize']=6,2
import warnings
warnings.filterwarnings('ignore')
```

```python
# Univariant analysis-->Plot graph with 1 col
vis1=sns.distplot(df["InternetUsers"])
plt.show()
```

```python
vis2=sns.displot(df["InternetUsers"])
plt.show()
```

```python
vis3=sns.distplot(df["InternetUsers"],bins=10)
plt.show()
```

# lower horizontal line==lower
# higher horizontal line==high
# Middle horizontal line==mean
# btw Middle and low horizontal line==25 percentile
# btw Middle and high  horizontal line==75 percentile
# outlier which is not near to the other values.Its very dangerous 

```python
# Bi Variant Analysis
vis4=sns.boxplot(data=df,x='IncomeGroup',y='BirthRate',patch_artist=True)
plt.rcParams['figure.figsize']=10,5
plt.show()
```

```python
df.columns
```

```python
vis4 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate')
plt.show()
```

```python
vis5=sns.lmplot(data=df,x='InternetUsers',y='BirthRate',fit_reg=False)   # fit_reg is for horizontal grpah line by default its true
plt.show()
```

```python
vis6 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup') #hue - parameter for color
plt.show()
```
