# Dot product in numpy arrays
If a and b are 1D arrays, it performs the standard dot product (scalar multiplication).
If a and b are 2D arrays, it performs matrix multiplication.
If one is a 1D array and the other is a 2D array, it performs matrix-vector multiplication.



```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = np.dot(a, b)
print(result)
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

result = np.dot(A, B)
result1=A@B  # Dot product can also be calculatred by using @ symbol
print(result1)

print(result)
```

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([7, 8, 9]).reshape(3,1)

result = np.dot(A, b)
print(result)
```

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([7, 8, 9])

result = np.dot(A, b)
print(result)

```
